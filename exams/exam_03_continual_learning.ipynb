### CC10. Accuracy matrix builder
Given a function `evaluate(task_id)->float` and T tasks, build the full T×T accuracy matrix (None for not-yet-evaluated entries). Only fill in row `t` entries 0..t.

```python
def build_acc_matrix(T, evaluate_fn):
    R=[[None]*T for _ in range(T)]
    for t in range(T):
        for i in range(t+1):
            R[t][i]=evaluate_fn(t,i)
    return R
```
